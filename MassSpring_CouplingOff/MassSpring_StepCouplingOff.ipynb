### Constant step function

The coupling between the two masses in the classical mass-spring system is the spring with spring constant $K$ between the two masses. This is the equivalent of the pump field in EIT. 

If I were to "turn off" the coupling between the two masses, then I'd have to remove the spring then put it back after some time. If I do this suddenly, then I can add a time dependence to $K(t)$ modelled by a step function:

$$
K(t) = \left\{\begin{array}{ll}
{K,} & {0 \leq t \leq 100} \\
{0,} & {100 < t\leq 200}\\
{K,} & {t > 200}
\end{array}\right.
$$

I kind of just arbitrarily chose to turn off the coupling at these times. I can probably switch them up to experiment with longer times to decouple the system.

A few definitions:

$$
\begin{aligned}
\Omega_r^2 &= K/m\\
\omega_0^2 &= k/m\\
\omega^2 &= (k + K)/m = \omega_0^2 + \Omega_r^2
\end{aligned}
$$


The differential equations for the coupled system is given by:

$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\omega^{2} x_{1}(t)-\Omega_{r}^{2} x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\omega^{2} x_{2}(t)-\Omega_{r}^{2} x_{1}(t)=0
$$

Make the following substitutions to the ODEs:
$\omega \rightarrow (k + K)/m$, $\Omega_r \rightarrow K/m$


$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\frac{k+K(t)}{m} x_{1}(t)-\frac{K(t)}{m} x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\frac{k+K(t)}{m} x_{2}(t)-\frac{K(t)}{m} x_{1}(t)=0
$$

The differential equations above are now kinda weird since there is a time dependence with $K(t)$. We'll get three different cases for certain time intervals for the ODEs:


1.) $0 \leq t \leq 100$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\frac{k+K}{m} x_{1}(t)-\frac{K}{m} x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\frac{k+K}{m} x_{2}(t)-\frac{K}{m} x_{1}(t)=0
$$

2.) $100 < t\leq 200$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\frac{k}{m} x_{1}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\frac{k}{m} x_{2}(t)=0
$$

3.) $t >200$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\frac{k+K}{m} x_{1}(t)-\frac{K}{m} x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\frac{k+K}{m} x_{2}(t)-\frac{K}{m} x_{1}(t)=0
$$

Looking at the second case, the system is completely decoupled (as expected!) and we can treat the two masses as two separate harmonic oscillators.

From cases 1 and 3, we should get back the same set of ODEs. So I'll try to incorporate the changing constant in time since it is known. 


I'll rewrite the three different cases in this notation:

1.) $0 \leq t \leq 100$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+(\omega_0^2 + \Omega_r^2) x_{1}(t)-\Omega_r^2 x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+(\omega_0^2 + \Omega_r^2) x_{2}(t)-\Omega_r^2 x_{1}(t)=0
$$

2.) $100 < t\leq 200$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+\omega_0^2 x_{1}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+\omega_0^2 x_{2}(t)=0
$$

3.) $t>200$
$$
\ddot{x}_{1}(t)+\gamma_{1} \dot{x}_{1}(t)+(\omega_0^2 + \Omega_r^2) x_{1}(t)-\Omega_r^2 x_{2}(t)=\frac{F}{m} e^{-i \omega_{s} t}
$$

$$
\ddot{x}_{2}(t)+\gamma_{2} \dot{x}_{2}(t)+(\omega_0^2 + \Omega_r^2) x_{2}(t)-\Omega_r^2 x_{1}(t)=0
$$

```python
# Most of the code is taken from MassSpringEITTransient.py with a few changes

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import scipy.integrate as sci
```

```python
# Some definitions
# frequencies are with respect to \omega = \sqrt[\omega1^2 + OmegaR^2]

gamma1 = 0.01
gamma2 = 0.01
omega0 = 0.91
OmegaR = np.sqrt(1-omega0**2)
Foverm = 1
```

```python
# Define a function Omega_r that models the step function
def Omega_r(omega0,t):
    if t > 100 and t <= 200:
        Omega_R = 0
    else:
        Omega_R = np.sqrt(1-omega0**2)
    return Omega_R

# Define function for coupled ODEs
def CoupledMassODE(t,y,p):
    # y[0] = x1: position of driven mass
    # y[1] = x2: position of undriven mass
    # y[2]= d(x1)/ dt
    # y[3]= d(x2)/ dt
    # parameters, p
    # p = (omega0, gamma1,gamma2,F,omegas)
    out = [y[2],
           y[3],
          -p[1]*y[2] - (p[0]**2 + (Omega_r(p[0],t))**2)*y[0] + (Omega_r(p[0],t)**2)*y[1] + p[3]*np.cos(p[4]*t),
          -p[2]*y[3] - (p[0]**2 + (Omega_r(p[0],t))**2)*y[1] + (Omega_r(p[0],t)**2)*y[0]]
    return out

# *** Define parameters and ICs
params = [omega0,gamma1,gamma2,Foverm,1]
#y0 = [0+0j,0+0j,0+0j,0+0j]
y0 = [0,-Foverm,0,0] # without complex drive, use real ICs
# **************************************** 

# *** Solve ODEs
sol = sci.solve_ivp(lambda t,y: CoupledMassODE(t,y,params),(0,500),y0)

# Steady State solutions
# When t->inf then OmegaR should take on OmegaR = np.sqrt(1-omega0**2)

# ******* x1: first mass dispacement ******* 
def x10single(omegas,omega0,gamma1,gamma2,F):
    # displacement of first mass
    # returns two arguments: amplitude and phase
    OmegaR = np.sqrt(1-omega0**2)
    x1 = ((1 - omegas**2 - 1j*gamma2*omegas)*F / 
            ((1 - omegas**2 - 1j*gamma1*omegas)*(1 - omegas**2 - 1j*gamma2*omegas)-OmegaR**4))
    return np.abs(x1),np.angle(x1)

x10 = np.frompyfunc(x10single,5,2)
# **************************************** 
      
      
# ******* x2: second mass dispacement ******* 
def x20single(omegas,omega0,gamma1,gamma2):
    # displacement of second mass
    # returns two arguments: amplitude and phase
    OmegaR = np.sqrt(1-omega0**2)
    x2= (OmegaR**2*Foverm / 
            ((1 - omegas**2 - 1j*gamma1*omegas)*(1 - omegas**2 - 1j*gamma2*omegas)-OmegaR**4))
    return np.abs(x2),np.angle(x2)

x20 = np.frompyfunc(x20single,4,2)
# **************************************** 


drives = np.linspace(0.1,2,2000)
x1sA,x1sph = x10(drives,omega0,gamma1,gamma2,Foverm)
x2sA,x2sph = x20(drives,omega0,gamma1,gamma2)   


# Just to know when the coupling is turned off in the plot, plot the coupling strength
# I will be scaling so the strength of the coupling is not accurately reflected in the plots.

times = np.array([0,100,200,np.max(sol.t)])
Kx1 = np.array([np.max(sol.y[0]),np.max(sol.y[0]),0,np.max(sol.y[0])])
Kx2 = np.array([np.max(sol.y[1]),np.max(sol.y[1]),0,np.max(sol.y[1])])

# ******* Plotting ******************************* 
# Steady states
f1 = plt.figure(1,figsize = (6,4))
plt.clf()
gs = gridspec.GridSpec(2,2,top = 0.80,left = 0.08,right = 0.95,hspace = 0,wspace = 0)
f1.suptitle('Steady-state solutions\n $\omega_1 =$ {:.2}, $\Omega_R$ = {:.2}, $\gamma_1 =$ {:.2}, $\gamma_2 =$ {:.2}'.format(omega0,OmegaR,gamma1,gamma2) )
#
# x1, amplitude
ax1 = plt.subplot(gs[0])
ax1.plot(drives, x1sA, 'k--')
ax1.set_title('Mass 1')
ax1.tick_params(axis = 'both',width=1, length = 3,direction = 'in',top='true',right = 'true')
ax1.set_xticklabels([])
ax1.set_ylabel(r'amplitude: $F/m\omega^2$')
ylims1 =ax1.get_ylim()

# x1, phase
ax3 = plt.subplot(gs[2])
ax3.plot(drives, x1sph, 'k--')
ax3.tick_params(axis = 'both',width=1, length = 3,direction = 'in',top='true',right = 'true')
ax3.set_ylim(-3.5,3.5)
ax3.set_xlabel(r'drive: $\omega_{\rm s}/\omega$')
ax3.set_ylabel(r'phase (rad)')


# x2 amplitude
ax2 = plt.subplot(gs[1])
ax2.plot(drives, x2sA, 'k--')
ax2.set_title('Mass 2')
ax2.tick_params(axis = 'both',width=1, length = 3,direction = 'in',top='true',right = 'true')
ax2.set_yticklabels([])
ax2.set_xticklabels([])
ax2.set_ylim(ylims1)

# x2, phase
ax4 = plt.subplot(gs[3])
ax4.plot(drives, x2sph, 'k--')
ax4.set_ylim(-3.5,3.5)
ax4.set_yticklabels([])
ax4.tick_params(axis = 'both',width=1, length = 3,direction = 'in',top='true',right = 'true')
ax4.set_xlabel(r'drive: $\omega_{\rm s}/\omega$')

# *** Plotting
# Transient states
f2 = plt.figure(2,figsize = (6,4))
plt.clf()
f2.suptitle('Transient Solutions \n $\omega_1 =$ {:.2},$\gamma_1 =$ {:.2}, $\gamma_2 =$ {:.2}'.format(omega0,gamma1,gamma2) )
gs1 = gridspec.GridSpec(2,1,top = 0.85,left = 0.08,right = 0.95,hspace = 0,wspace = 0)

# Driven mass position
ax1 = plt.subplot(gs1[0])
#ax1.plot(times,x1ss,'k-',alpha = 0.5,label = 'x1, steady state')
ax1.plot(times,Kx1,drawstyle='steps',label='coupling')
ax1.plot(sol.t,sol.y[0],'r-',label = 'x1, transient')
ax1.legend(loc=4)
ax1.set_ylabel(r'amplitude: $F/m\omega^2$')

# Passive mass position
ax2 = plt.subplot(gs1[1])
#ax2.plot(times,x2ss,'k-',alpha = 0.5,label = 'x2, steady state')
ax2.plot(times,Kx2,drawstyle='steps',label='coupling')
ax2.plot(sol.t,sol.y[1],'b-',label = 'x2, transient')
ax2.legend(loc =4)
ax2.set_ylabel(r'amplitude: $F/m\omega^2$')
ax2.set_xlabel(r'time: $1/\omega$')
```
